```python
import numpy as np
import matplotlib.pyplot as plt

T = int(input("baze zamani ra vared konid : "))
t = [i for i in range(T)]
B = [100 for i in range(T)]
for i in range(10):
  for i in range(1,T):
    X = np.random.normal(0 ,t[i] - t[i-1])
    B[i] = B[i - 1] + X
  plt.plot(t , B)
plt.show()
```

```python
from numpy import *
import matplotlib.pyplot as plt

paths = 100 #int(input("تعداد شبیه سازی : "))
initial_price = 100 #float(input("قیمت اولیه سهم : "))
volatility = 0.3 #float(input("نوسان قیمت : "))
t = 365 #int(input(" بازه زمانی مورد نظر : ماهانه(12)، هفتگی(52) و یا روزانه (365) : "))
drift = 0.08 #float(input("راندگی : "))

prices = list()
current_price = initial_price
dt = 1 / t
T = 1
dYt = 0
price_paths = []
while T - dt > 0 :
  dWt = random.normal(0 ,sqrt(dt))  # حرکت براونی
  dYt = drift * dt + volatility * dWt  # تغییرات قیمت
  current_price += dYt  # اعمال تغییرات بر روی قیمت غعلی
  prices.append(current_price)  # اضافه کردن قیمت جدید
  T -= dt  # شرط برقراری حلقه
for i in range(0, paths):
  price_paths.append(prices) #اندازه قیمت در هر نقطه از مسیر 
for i in price_paths:
  plt.plot(i)
plt.show()
print(price_paths ,'\n')
print(dYt ,'\n')
print(dWt ,'\n')
print(current_price ,'\n')
print(prices ,'\n')
```

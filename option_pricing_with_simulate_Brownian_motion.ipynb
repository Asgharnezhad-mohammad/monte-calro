```python
from numpy import *
import matplotlib.pyplot as plt

paths = 100 #int(input("   : "))
initial_price = 100 #float(input("   : "))
volatility = 0.3 #float(input("  : "))
t = 365 #int(input("     : (12) (52)    (365) : "))
drift = 0.08 #float(input(" : "))

prices = list()
current_price = initial_price
dt = 1 / t
T = 1
dYt = 0
price_paths = []
while T - dt > 0 :
  dWt = random.normal(0 ,sqrt(dt))  #  
  dYt = drift * dt + volatility * dWt  #  
  current_price += dYt  #      
  prices.append(current_price)  #    
  T -= dt  #   
for i in range(0, paths):
  price_paths.append(prices) #       
for i in price_paths:
  plt.plot(i)
plt.show()
print(price_paths ,'\n')
print(dYt ,'\n')
print(dWt ,'\n')
print(current_price ,'\n')
print(prices ,'\n')
```
